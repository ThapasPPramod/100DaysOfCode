```python
import random, requests
from bs4 import BeautifulSoup

def scrape_website(url):
  try:
    response = requests.get(url)

    if response.status_code == 200:
      soup = BeautifulSoup(markup=response.text,features='html.parser' )
      print('soup: ', soup)
      text_content = soup.get_text()
      print(text_content)
      return text_content
    else:
      print(f'Error: {response.status_code}')
  except Exception as e:
    print(f"Error :{e} ")
    exit(0)

website_url = 'https://en.wikipedia.org/wiki/Hangman_(game)'
text_content = scrape_website(website_url)
```

```python
list_ = list(set(text_content.split()))
```

```python
import re, nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
def clean(string, stemming):
  words = re.sub('[^a-zA-Z]',' ', string)
  words = words.lower()
  #stemming
  words = words.split()
  print(words)
  sw = stopwords.words('english')
  print(type(sw))
  ps = PorterStemmer()
  string = [word for word in words if len(word)>3 and len(word)<9 and (not stemming or word not in set(sw))]
  return string
stemmed_words = clean(' '.join(list_), False)
```

```python
stemmed_words
```

```python
from nltk import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize

tokenized_words = word_tokenize(" ".join(stemmed_words))
pos_tags = pos_tag(tokenized_words)
nouns = [word for word, pos in pos_tags if pos.startswith('N')]
print(nouns)
```

```python
class hangman():
  def __init__(self):
    self.word = self.getRandomWord()
    self.unused = list(self.word)
    self.len = len(self.word)
    self.vis = ['_'] * self.len
    self.vis2 = 'Hangman?'
    self.stage = -1
    self.gameover = False
    self.used = set()
    self.glen = 0
    self.status = 'You lose!'
    self.body= [' '] * 6
  def updateMan(self):
    human = ['O', '|', '\\','/','/','\\']
    self.body[self.stage] = human[self.stage]
    return

  def getRandomWord(self):
    try:
      word = random.choice(nouns)
    except Exception as e:
      print(f'Error :{e}')
      exit(0)
    return word
  def display(self):
    # print('-'*10)
    # print('|',' '*10,'|')
    print('T',' ',self.body[0],' ',self.vis2)
    print('|',self.body[2], self.body[1], self.body[3],self.vis)
    print('|','', self.body[4], self.body[5])
    # rint('|',' '*10,'|')p
    return
  def update(self, c):
    if c in self.used:
      self.vis2 = 'you guessed this already! ' + f'You used {self.used}'
    elif c in self.unused:
      self.used.add(c)
      self.vis2 = f'You used {self.used}'
      while True:
        try:
          index = self.unused.index(c)
          self.vis[index] = c
          self.unused[index] = '_'
          self.glen += 1
        except Exception as e:
          print(f'{c} found !')
          if self.glen == self.len:
            self.gameover = True
            self.status = 'You won!'
          break
    else:
      self.stage += 1
      self.updateMan()
      if self.stage == 5:
        self.gameover = True
        self.vis = f'Game Over ! The word was {self.word} !'
      self.used.add(c)
      self.vis2 = f'You used {self.used}'



    return
```

```python
from builtins import input
game = hangman()
while True:
  game.display()
  if game.gameover:
    print(game.status)
    break
  user_input = input('enter char :').strip()
  game.update(user_input)
```
